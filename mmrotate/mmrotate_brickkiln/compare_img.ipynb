```python
import os
import random
from PIL import Image
random.seed(42)  # For reproducibility

import matplotlib.pyplot as plt

def display_random_images(dir1, dir2, num_images=10):
    # List all image files in dir1 (assuming both dirs have same image names)
    image_files = [f for f in os.listdir(dir1) if os.path.isfile(os.path.join(dir1, f))]
    # Randomly select images
    selected_files = random.sample(image_files, min(num_images, len(image_files)))
    
    plt.figure(figsize=(20, 4 * num_images))
    for idx, img_name in enumerate(selected_files):
        img1_path = os.path.join(dir1, img_name)
        img2_path = os.path.join(dir2, img_name)
        
        # Display image from dir1
        plt.subplot(num_images, 2, idx*2 + 1)
        img1 = Image.open(img1_path)
        plt.imshow(img1)
        plt.title(f"{img_name} (Dir1)")
        plt.axis('off')
        
        # Display image from dir2
        plt.subplot(num_images, 2, idx*2 + 2)
        img2 = Image.open(img2_path)
        plt.imshow(img2)
        plt.title(f"{img_name} (Dir2)")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Example usage:
# display_random_images('/path/to/dir1', '/path/to/dir2')
```
