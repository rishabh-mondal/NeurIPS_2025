```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
history_path = "runs/sentinel_data/history.csv"

df = pd.read_csv(history_path)
# display(df.head())
```

```python
# columns in df
df.columns.tolist()
```

```python
plt.figure(figsize=(12, 8))

plt.plot(
	df['epoch'],
	df['train_loss'],
	label='Training Loss',
	marker='o',
	linestyle='-'
)

plt.plot(
	df['epoch'],
	df['test_loss'],
	label='Validation Loss',
	marker='o',
	linestyle='--'
)

plt.title('Training and Validation Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# plt.savefig('train_validation_loss.png')
plt.show()
```
